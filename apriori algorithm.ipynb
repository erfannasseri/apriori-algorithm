```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
df = pd.read_csv('dataset.csv')
df
```

```python
df.drop('date_time',axis=1 , inplace=True)
df.drop('period_day',axis=1 , inplace=True)
df.drop('weekday_weekend',axis=1 , inplace=True)
df
```

```python
pd =[]
tr =[]
it =[]
for i in range(20507):
    condition = df['Transaction'] == i   
    tr = tr + [i]
    d = df[condition]
    pd = list(d["Item"].apply(lambda x:x))
    it = it + [pd]
    
#print(tr)   
#print(it)  
```

```python
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
```

```python
a_data = a.fit(it).transform(it)
a_data
```

```python
 a.columns_
```

```python
import pandas as pd
df = pd.DataFrame(a_data, columns=a.columns_)
df = df.replace(False,0)
df
```

```python
df = apriori(df,min_support=0.017, use_colnames= True , verbose=1)
df
```

```python
df_ar = association_rules(df, metric="confidence",min_threshold=0.01)
df_ar
```
